```python
import numpy as np
from scipy.interpolate import griddata
import matplotlib.pyplot as plt
%matplotlib inline
data = numpy.loadtxt("testing.dat");
x = data[:,0]
y = data[:,1]
z = data[:,2]

def plot_contour(x,y,z,resolution = 1000,contour_method='linear'):
    resolution = str(resolution)+'j'
    X,Y = np.mgrid[min(x):max(x):complex(resolution),   min(y):max(y):complex(resolution)]
    points = [[a,b] for a,b in zip(x,y)]
    Z = griddata(points, z, (X, Y), method=contour_method)
    return X,Y,Z

X,Y,Z = plot_contour(x,y,z,resolution = 1000,contour_method='linear')

with plt.style.context("seaborn-white"):
    fig, ax = plt.subplots(figsize=(13,8))
  #  ax.scatter(x,y, color="black", linewidth=1, edgecolor="ivory", s=50)
    ax.contourf(X,Y,Z)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
import math
k=pd.read_csv('densmap.dat', sep='\s+', header=None, nrows=2 )
j=np.array(k)
a=int(j[0])+1
b=int(j[1])
z0=-238.588165
boxlen=10
#a number of slices
#b number of rows for each slice


for i in range(1,a):
    z1=z0+boxlen*(i-1)
    z2=z0+boxlen*i
    title=str('slice '+ str(z1) +' to ' + str(z2))
    f=plt.figure(i)
    f.suptitle(title, fontsize=20)
    indata=pd.read_csv('densmap.dat', sep='\s+', header=None,skiprows=3+((i-1)*(b+1)),nrows=b)
#    print(indata)
    densdata=indata.values 
    plt.imshow(densdata, cmap='coolwarm', interpolation='nearest')
    cb=plt.colorbar()
    plt.savefig(str(title+'.png'))
#    Image.open(str(title+'.png')).save(str(title+'.jpg','JPEG')
#    f.show()
```
